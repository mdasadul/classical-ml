# K-Nearest Neighbors Classification

The dataset was obtained from https://archive.ics.uci.edu/ml/datasets/Iris.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# read data and parse it
data = pd.read_csv('./datasets/iris_knn_classification.csv', 
                   names=["sepal_length","sepal_width","petal_length","petal_width","class_name"])

setosa = data[data['class_name'] == 'Iris-setosa']
setosa = setosa.loc[:, setosa.columns != 'class_name']

virginica = data[data['class_name'] == 'Iris-virginica']
virginica = virginica.loc[:, virginica.columns != 'class_name']

versicolor = data[data['class_name'] == 'Iris-versicolor']
versicolor = versicolor.loc[:, versicolor.columns != 'class_name']
```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)
fig.suptitle('Scatter plots of different flower features')

# Sepal Length vs Sepal Width
ax = fig.add_subplot(231)
ax.scatter(setosa.sepal_length, setosa.sepal_width, color='green', marker='.', label='Sentosa')
ax.scatter(virginica.sepal_length, virginica.sepal_width, color='#222222', marker='+', label='Virginica')
ax.scatter(versicolor.sepal_length, versicolor.sepal_width, color='red', marker='*', label='Versicolor')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_title('Sepal Length vs Sepal Width')
plt.legend(loc='lower right')

# Petal Length vs Petal Width
ax = fig.add_subplot(232)
ax.scatter(setosa.petal_length, setosa.petal_width, color='green', marker='.', label='Sentosa')
ax.scatter(virginica.petal_length, virginica.petal_width, color='#222222', marker='+', label='Virginica')
ax.scatter(versicolor.petal_length, versicolor.petal_width, color='red', marker='*', label='Versicolor')
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Petal Length vs Petal Width')
plt.legend(loc='lower right')

# Sepal Length vs Petal Width
ax = fig.add_subplot(233)
ax.scatter(setosa.sepal_length, setosa.petal_width, color='green', marker='.', label='Sentosa')
ax.scatter(virginica.sepal_length, virginica.petal_width, color='#222222', marker='+', label='Virginica')
ax.scatter(versicolor.sepal_length, versicolor.petal_width, color='red', marker='*', label='Versicolor')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Sepal Length vs Petal Width')
plt.legend(loc='lower right')

# Petal Length vs Sepal Width
ax = fig.add_subplot(234)
ax.scatter(setosa.petal_length, setosa.sepal_width, color='green', marker='.', label='Sentosa')
ax.scatter(virginica.petal_length, virginica.sepal_width, color='#222222', marker='+', label='Virginica')
ax.scatter(versicolor.petal_length, versicolor.sepal_width, color='red', marker='*', label='Versicolor')
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_title('Petal Length vs Sepal Width')
plt.legend(loc='upper right')

# Sepal Length vs Petal Length
ax = fig.add_subplot(235)
ax.scatter(setosa.sepal_length, setosa.petal_length, color='green', marker='.', label='Sentosa')
ax.scatter(virginica.sepal_length, virginica.petal_length, color='#222222', marker='+', label='Virginica')
ax.scatter(versicolor.sepal_length, versicolor.petal_length, color='red', marker='*', label='Versicolor')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Petal Length (cm)')
ax.set_title('Sepal Length vs Petal Length')
plt.legend(loc='lower right')

# Sepal Width vs Petal Width
ax = fig.add_subplot(236)
ax.scatter(setosa.sepal_width, setosa.petal_width, color='green', marker='.', label='Sentosa')
ax.scatter(virginica.sepal_width, virginica.petal_width, color='#222222', marker='+', label='Virginica')
ax.scatter(versicolor.sepal_width, versicolor.petal_width, color='red', marker='*', label='Versicolor')
ax.set_xlabel('Sepal Width (cm)')
ax.set_ylabel('Petal Width (cm)')
ax.set_title('Sepal Width vs Petal Width')
plt.legend(loc='center right')
```
